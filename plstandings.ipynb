```python
#now lets create a plot 
bumpy = Bumpy(
    scatter_color = '#282828', line_color = 'black',
    rotate_xticks=90,
    ticklabel_size=28,
    scatter_primary='D',
    show_right=True,
    plot_labels=True,
    alignment_yvalue=.1,
    alignment_xvalue=.065
)
```

```python
#copying the original data into another one just to avoid wrong computation on original data
df_copy = df.copy()
```

```python
# we only need season number, team name and its position in corresponding seasons
df_copy = df[["Season","Team","Pos"]]
```

```python
df_copy = df_copy['Season'].replace("2000-01","Season 1",inplace=True)
#likewise we need to replace each season with such names
```

```python
df['Se'] = df['Season'].replace(to_replace="2000-01",value="HEY",inplace=True)
```

```python
# Copying the edited original data into a new variable
new_df = df_copy.copy()
pd.DataFrame(new_df)
```

```python
#calculating no of occurance of each team in the dataset
vc = new_df['Team'].value_counts()
```

```python
#lets select teams that have minimum count as 5
mask = new_df['Team'].isin(vc[vc >= 5].index)

# Filter the DataFrame using the mask to keep only the rows with sufficient counts
filtered_df = new_df[mask]
```

```python
filtered_df
```

```python
#to get appropriate results, we need to transform the data
#setting season number as index and team names as columns with Position as values will tell us each teams position in each season
#this will help us plot the data better
pivot_df = pd.pivot_table(filtered_df, index='Season', columns='Team', values='Pos', aggfunc='first')
```

```python
#picking only 8 big teams : Arsenal,Chelsea,Everton,Liverpool,Manchester United,Tottenham Hotspur,Manchester City,Newcastle United
pivot_df = pivot_df[["Arsenal","Chelsea","Everton","Liverpool","Manchester United","Tottenham Hotspur","Manchester City","Newcastle United"]]
```

```python
pivot_df
```

```python
#now lets plot this data and see which team improves over season and which team needs to improve
fig,ax = bumpy.plot(
   x_list = pivot_df.index, #text to be plotted on x-axis
    y_list = np.linspace(1,20,20).astype(int),#placing sequence of numbers from 1 to 20 denoting positions
    values = pivot_df, #data which contains key as team name and value as position 
    secondary_alpha = .5,
    highlight_dict = highlight_dict,#used to denote user defined colors for each team
    figsize = (50,20),
    y_label = 'Position throughout seasons',
    ylim = (-.1,22),#no of y axes and x axes
    lw = 2.5 #line width
)
plt.grid(True) #sets gridlines as true
fig.text(s = 'BIG TEAMS PERFORMANCE FROM SEASON 1 TO 22',x = .30,y=.90,c = 'white',size= 54,weight = 'bold') #title
highlight_text.fig_text(x = .25,y=.88, #helps us position the subtitle
                       s = 'Comparing <MANCHESTER UNITED>, <MANCHESTER CITY>, <ARSENAL>, <CHELSEA>, <TOTTENHAM HOTSPUR>, <NEWCASTLE UNITED>, <LIVERPOOL>, & <EVERTON>'
                        ,c = 'white', fontsize = 25)#subtitle


ax2 = fig.add_axes([.075,.85,.12,.12])#helps us position the logo and the size
ax2.axis("ON")
img = Image.open('Premier-League-Logo-PNG-Image.png')
ax2.imshow(img)
```
